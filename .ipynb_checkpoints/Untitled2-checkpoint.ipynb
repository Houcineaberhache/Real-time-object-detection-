Q4. Tracez le nuage de points (plt.scatter) en utilisant des couleurs différentes pour Accepted (0 ou 1), en vous
basant sur les données d'entraînement (train set). Une simple ligne droite (comme dans le TP3) peut-elle séparer
efficacement ces deux groupes ?

```python
plt.figure(figsize=(8,6))
plt.scatter(X_train[y_train==1,0],X_train[y_train==1,1],color='red',label='accepted')
plt.scatter(X_train[y_train==0,0],X_train[y_train==0,1],color='blue',label='rejected')
plt.xlabel('test1')
plt.ylabel('test2')
plt.legend()
plt.show()
```

Non une ligne droit ne peut pas separer les deux groupes
